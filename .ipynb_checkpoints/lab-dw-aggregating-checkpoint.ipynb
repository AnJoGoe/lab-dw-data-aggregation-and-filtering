## 2. Using the original Dataframe, analyze the average total_claim_amount by each policy type and gender for customers who have responded "Yes" to the last marketing campaign. Write your conclusions.

```python
dfm[dfm['response'] == "Yes"]
```

```python
dfm[dfm['response'] == "Yes"].groupby(['policy_type', 'gender'])['total_claim_amount'].mean()
```

## 3. Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.

```python
dfm['state'].unique()
```

```python
# total number of customers who have policies in each state

customer_state = dfm.groupby(['state'])['customer'].count()
customer_state
```

```python
dfm.shape
```

```python
# filter the results to only include states where there are more than 500 customers
customer_state_500 = customer_state[customer_state > 500]
customer_state_500
```

## 4. Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.

```python
order_list = ['Doctor', 'Master', 'Bachelor', 'College', 'High School or Below']

df_life_edu_gen = dfm.groupby(["gender", "education"]).agg({"customer_lifetime_value": ['min','max','median']})

df_life_edu_gen.reindex(pd.MultiIndex.from_product([df_life_edu_gen.index.levels[0], order_list], names=df_life_edu_gen.index.names))
```

```python
dfm['education'].unique()
```

<b style="color:green"> Conclusion </b>

There seems NOT to be an impact of the gender variable on the "customer_life_time" value. But the "customer_life_time_value" seems to decrease with higher degree of educational level. So customers with e lower educational level might generate more revenue over the time. But the data is highly variable with big range between min and max and tend to have extreme values in the higher ranges (the median is closer to the min value).
